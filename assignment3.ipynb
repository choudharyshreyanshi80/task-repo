```python
# Python function to find the maximum of three numbers.

def func(num1,num2,num3):
    if num1>num2:
        if  num1>num3:
            return num1
        else:
            return num3
    elif num2>num3:
        return num2
    else:
        return num3

func(23,65,54)
```

```python
# Python function that takes a list and returns a new list with distinct elements from the first list.

def func(first_list):
    new_list=set(first_list)
    return list(new_list)

func([3,2,3,5,4,5,1,1])
```

```python
# Python function to multiply all the numbers in a list.


def func(lst):
    num=1
    for x in lst:
        num=num*x
    return num

func([2,5,7,3])
```

```python
# Python function to calculate the factorial of a number.

def func(num):
    if num < 0:
        return "Factorial does not exist"
    
    fact=1
    for i in range(1,num+1):
        fact*=i
    return fact

func(8)
```

```python
#  Python function to check whether a number falls within a given range.

def func(num,start,end):
    if  num in range(start,end+1):
        return "within a given range"
    else:
        return "not in range"

func(6,0,6)
```

```python
# Python function to Print Even Numbers from a Given List.

def func(lst):
    for x in lst:
        if x % 2 == 0:
            print(x)
        
func([2,3,5,6,7,8])
```

```python
# Python function that takes a number as a parameter and checks whether the number is prime or not.

def func(num):
    if num <= 1:
        return "Not Prime"

    for i in range(2, num):
        if num % i == 0:
            return "Not Prime"
    return "Prime"

func(73)
```

```python
# Python function that accepts a string and counts the number of upper and lower case letters. 

def func(string):
    upper=0
    lower=0
    for x in string:
        if x.isupper():
            upper+=1
        elif x.islower():
            lower+=1
        
    print("Upper case letters:",upper)
    print("Lower case letters:",lower)

func("Hello!! Python")
```
