```python
stud_name=input("Enter name:")
stud_class=int(input("Enter class:"))
sub1=float(input("Subject1 marks:"))
sub2=float(input("Subject2 marks:"))
sub3=float(input("Subject3 marks:"))
sub4=float(input("Subject4 marks:"))
sub5=float(input("Subject5 marks:"))
total_marks=sub1+sub2+sub3+sub4+sub5
percentage=(total_marks/500)*100
print("Name:",stud_name)
print("Class:",stud_class)
print("Total marks are: ",total_marks)
print("Percentage:",percentage)
```
