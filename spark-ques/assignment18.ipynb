```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,udf
```

```python
spark=SparkSession.builder.appName("demoapp").getOrCreate()
```

```python
df=spark.range(0,5000000)
```

```python
df_partitions=df.rdd.getNumPartitions()
print("Before partitions:",df_partitions)
```

```python
df_repartitioned=df.repartition(12)
repartition_count=df_repartitioned.rdd.getNumPartitions()
print("After repartition:",repartition_count)
```

```python
df_coalesced=df_repartitioned.coalesce(3)
coalesced_count=df_coalesced.rdd.getNumPartitions()
print("After coalesce:",coalesced_count)
```

```python
spark.stop()
```
