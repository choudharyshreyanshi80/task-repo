```python
s={"CS","DS","AI",10}
set1={10,20,30}

for x in s:
    print(x)

s.add("EE")
set2=set1.union(s)
print(set2)

if set1>s:
    print("set1 is greater than s")
else:
    print("s is greater")

set3=set1.intersection(s)
print(set3)

set3.clear()
print(set3)
```

```python
num1=int(input("Enter number1:"))
num2=int(input("Enter number2:"))
def add(x,y):
    return x+y
def multiply(x,y):
    return x*y
def subtract(x,y):
    return x-y
def division(x,y):
    return x/y

print("Addition:",add(num1,num2))
print("Multiplication:",multiply(num1,num2))
print("Subtraction:",subtract(num1,num2))
print("Division:",division(num1,num2))
```

```python
num=int(input("Enter a number:"))

number=num
rev=0
while num>0:
    digit=num%10
    rev =rev*10 + digit
    num=num//10
if number==rev:
    print("It is a Palindrome number")
else:
    print("Not a Palindrome number")
```
